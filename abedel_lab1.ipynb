```python
# Exercise Set 1 (a)
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, n): # Could make this faster
    if n%i == 0:
      return False
  return True

# Test Function
print(is_prime(11))
```

```python
# Exercise Set 1 (b)
class Rectangle:
  def __init__(self, width=0, height=0):
    self.width = width
    self.height = height
  def area(self):
    return self.width * self.height * 1.0
  def perimeter(self):
    return 2.0 * (self.width + self.height)
  def diagonal(self):
    return np.sqrt(self.width**2 + self.height**2)

# Test Function
myRectangle = Rectangle(3, 4)
print("My rectangle has an area of %d, perimeter of %d, and diagonal of %d. " \
      % (myRectangle.area(), myRectangle.perimeter(), myRectangle.diagonal()))
```

```python
# Exercise Set 1 (c)
def create_grade_dict(names, grades):
  newDict = {}
  for name in names:
    newDict[name] = grades[names.index(name)]
  return newDict

# Test Function
names = ['Alice', 'Bob', 'Charlie']
grades = [85, 92, 78]
print(create_grade_dict(names, grades))
```

```python
# Exercise Set 1 (d)

# For 1-n, create empty str and update with 'Fizz' or 'Buzz' as necessary.
# If string remains empty, just print current number instead
def fizz_buzz(n):
  for i in range(1, n+1):
    string = ''
    if i % 3 == 0:
      string = 'Fizz'
    if i % 5 == 0:
      string = string + 'Buzz'
    print(string if string else i)

  # Test Function
fizz_buzz(15)
```

```python
# Exercise Set 1 (e)
def nested_average(data):
  newDict = {}
  for key in data.keys():
    newDict[key] = np.mean(data[key])
  return newDict

# Test Function
scores = {'Math': [90, 85, 88], 'Science': [78, 82, 91],'English': [85, 87, 89]}
print(nested_average(scores))
```

```python
# Exercise Set 2 (a)
url = 'https://raw.githubusercontent.com/aims-umich/ners590data/main/boston.csv'
data = pd.read_csv(url)
```

```python
# Exercise Set 2 (b)
print(data.head(5))
data.describe()
```

```python
# Exercise Set 2 (c)
print(data.iloc[:,2])
print('\n--------------------------------------------------\n')
print(data.iloc[99:105,:])
print('\n--------------------------------------------------\n')
print(data.iloc[99:105,4:7])
```

```python
# Exercise Set 2 (d)

# filter data by bool
filtered_data = data[data['MEDV'] > 30]

# Sort and print
sorted_homes = filtered_data.sort_values('RM')
print(sorted_homes.head(10))
```

```python
# Exercise Set 2 (e)

# If PRICE_PER_ROOM does not exist, create it
if 'PRICE_PER_ROOM' not in data.columns:
  data.insert(data.shape[1], 'PRICE_PER_ROOM', \
              data.loc[:, 'MEDV'] / data.loc[:, 'RM'])

print(data.iloc[0:10,:])
```

```python
# Exercise Set 2 (f)
plt.scatter(data.loc[:,'RM'], data.loc[:,'MEDV'])
plt.xlabel('Number of Rooms')
plt.ylabel('Median Value')
plt.show()
```

```python
# Exercise Set 2 (g)

# If HIGH_VALUE does not exist, create it
if 'HIGH_VALUE' not in data.columns:
  median_value = data.median().loc['MEDV']
  data.insert(data.shape[1], 'HIGH_VALUE', data.loc[:,'MEDV'] > median_value)

print(data.iloc[0:10,:])
```

```python
# Exercise Set 2 (h)
correlation_matrix = data.corr()

# To find which columns are most highly correlated with MEDV, just
# inspect the correlation matrix below (RM, LSTAT, and our own columns)

# I think a diverging color palette works better here
sns.heatmap(correlation_matrix, annot = True, fmt = '.1f') #, cmap = 'vlag', \
            #vmin = -1.0, vmax = 1.0)
```
